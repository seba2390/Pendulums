```python
import matplotlib.pyplot as plt
import numpy as np
from Spring import Spring
```

```python
# Initial points
start_point = np.array([0, 3])
end_point = np.array([3, 3])

spring = Spring(start=start_point, 
                end=end_point)

x, y = spring.get_points()

fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.plot(x, y)
ax.plot([start_point[0],end_point[0]],[start_point[1],end_point[1]], 'ro')
ax.set_xlim(-1,6)
ax.set_ylim(-1,6)
ax.set_aspect('equal')
```
